```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 10)

# step 1: import/build your stop word dictionary

# step 2: create a TfidfVectorizer and remove stop words
tfidf = TfidfVectorizer(stop_words='english')

df['combined'] = df['director'].str.replace(" ", "") + " " + df['description']

# step 3: fit and transform the data to a dfidf matrix
tfidf_matrix = tfidf.fit_transform(df['combined'])

# dataframe helps us view the results
df_tfidf = pd.DataFrame(tfidf_matrix.T.todense(), index=tfidf.get_feature_names_out(), columns=df['combined'])
df_tfidf
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

df_results = pd.DataFrame(cosine_sim, index=df['show_id'], columns=df['show_id'])
df_results.sort_values(by=['s8'], ascending=False)

# deployment is saving df_results as a .sav file and returning the top 5 (top 6 because the top recommendation is always itself)
```
