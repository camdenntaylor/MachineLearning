```python
# FOR ASSIGNMENT
import sqlite3
# creates database
conn = sqlite3.connect('nba.db')
```

```python
create_sql = "CREATE TABLE IF NOT EXISTS players (name TEXT, age INTEGER, team TEXT)"
# cursor can run sql commands on the database with cursor.execute
cursor = conn.cursor()
cursor.execute(create_sql)
```

```python
for row in df.itertuples():
    name = row[1].replace("'", "")
    insert_sql = f"INSERT INTO players (name, age, team) VALUES ('{name}', {row.Age}, '{row.Team}')"
    cursor.execute(insert_sql)
```

```python
#read from db
read_sql = "SELECT * FROM players"
table = cursor.execute(read_sql)
table.fetchall()
```
