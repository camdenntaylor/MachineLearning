```python
import pandas as pd
import statsmodels.api as sm
import sys
sys.path.append(r'C:\Users\camde\OneDrive - Brigham Young University\IS JUNIOR CORE-Camdens-Laptop\IS 455\myfunctions')
import myfunctions as my

# add days_since_release column
df = pd.read_csv(r'C:\Users\camde\OneDrive - Brigham Young University\IS JUNIOR CORE-Camdens-Laptop\IS 455\modeling\disney_movies_total_gross.csv')
df['days_since_release'] = (pd.to_datetime("01/01/2017") - pd.to_datetime(df['release_date'])).dt.days

# create regression  model from days_since_release and total_gross
y = df['total_gross']
X = df[['days_since_release']].assign(const=1)  # Add const=1 because we're in statsmodels

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python

# Fill in all missing values of mpaa_rating with the value "Empty
df.loc[df['mpaa_rating'].isna(), 'mpaa_rating'] = 'Empty'

# Create dummy code features to convert mpaa_rating to sets of numeric 0/1 features for all values except "Empty"
dummies = pd.get_dummies(df['mpaa_rating'], drop_first=True)
dummies = dummies.astype(int)
df = pd.concat([df.drop(columns=['mpaa_rating']), dummies], axis=1)


df.head(20)
```

```python
# Run another MLR a few rows below the previous one using both days_since_release and all of the dummy codes you just created for mpaa_rating
y = df['total_gross']
X = df[['days_since_release', 'G', 'Not Rated', 'PG', 'PG-13', 'R']].assign(const=1)  # Add const=1 because we're in statsmodels

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Fill in the missing values of genre with "Not Specified".
df.loc[df['genre'].isna(), 'genre'] = 'Not Specified'
```

```python
# Create dummy codes for all 12 values of genre except for "Not Specified".
dummies = pd.get_dummies(df['genre'])
dummies = dummies.astype(int)
df = pd.concat([df.drop(columns=['genre']), dummies], axis=1)


# df.head(20)
df.columns
```

```python
# Generate an MLR based using these 12 columns as well as days_since_release as well as the mpaa_rating columns for 
# G, PG, and PG-13 movies. In other words, do not include the dummy codes for R and Not Rated movies. This should give
#  you a total of 16 features (not including the label).
df.drop(columns=['Not Specified'], inplace=True)
y = df['total_gross']
X = df[['days_since_release', 'G', 'PG', 'PG-13', 'Action', 'Adventure', 'Black Comedy', 
        'Comedy', 'Concert/Performance', 'Documentary', 'Drama', 'Horror', 'Musical', 
        'Romantic Comedy', 'Thriller/Suspense', 'Western']].assign(const=1)  # Add const=1 because we're in statsmodels

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Rerun the same set of features in your last model, but substitute inflation_adjusted_gross as the new label instead of total_gross
y = df['inflation_adjusted_gross']
X = df[['days_since_release', 'G', 'PG', 'PG-13', 'Action', 'Adventure', 'Black Comedy', 
        'Comedy', 'Concert/Performance', 'Documentary', 'Drama', 'Horror', 'Musical', 
        'Romantic Comedy', 'Thriller/Suspense', 'Western']].assign(const=1)  # Add const=1 because we're in statsmodels

model = sm.OLS(y, X).fit()
print(model.summary())
```
