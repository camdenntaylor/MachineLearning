```python
# Question 8: Drop any rows that contain null values in any column using .dropna(). Use the statsmodels package to create an MLR/OLS model to predict math score as a function of the reading score and writing score. Do not split the data; use the entire dataset for training. Allow the y-intercept to vary.
df_students.dropna(inplace=True)

import statsmodels.api as sm
y = df_students['math score']
X = df_students[['reading score', 'writing score']].assign(const=1)  # Add const=1 because we're in statsmodels

model = sm.OLS(y, X).fit()
print(model.summary())
# Which feature best predicts their math score?
# reading score
```

```python
#  Question 18: Import all needed packages for classification modeling, splitting the data, and calculating fit metrics. 
# Import the Lending Club dataset lc_xs.csv included with this midterm. 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt

loans_df = pd.read_csv('lc_xs.csv')
display(loans_df.head())
print(loans_df.shape[0])
# How many records are in the dataset?

```

```python
# Question 19: Run the code below. If you named your DataFrame something other than 'df', then be sure to change 'df' 
# to whatever your DataFrame is named before running the code.
loans_df['loan_status'].value_counts()
# Based on the results, do you believe that Lending Club is doing a decent job of determining who will pay back their loans?

```

```python
# QUESTION 20: Drop the features: loan_status_numeric, issue_d, title, emp_title, emp_length, earliest_cr_line, mths_since_last_delinq, 
# mths_since_last_record, sub_grade. Pseudocode:
#      df.drop(columns=[name, name, ..., name], inplace=True)
loans_df.drop(columns=['loan_status_numeric', 'issue_d', 'title', 'emp_title', 'emp_length', 'earliest_cr_line', 'mths_since_last_delinq', 
                       'mths_since_last_record', 'sub_grade'], inplace=True)

# Since customers who are Current, In Grace Period, Late (16-30 days), and Late (31-120 days) have not finished their loan, 
# drop all records except for those which are Fully Paid or Charged Off. Pseudocode: 
#      df.loc[(df['column_name'] == A) | (df['column_name'] == B)]
loans_df = loans_df[loans_df['loan_status'].isin(['Fully Paid', 'Charged Off'])]

# Drop any rows containing null data using the command: df.dropna(inplace=True). 
loans_df.dropna(inplace=True)

# Finally, create dummy codes for all remaining categorical features (but not for the label loan_status)
X = pd.get_dummies(loans_df.drop(columns=['loan_status']), drop_first=True, dtype=int)

# Print out the first five records of the resulting DataFrame. How many columns are left after completing each of these tasks?
X.head()
```

```python
# Question 21: Set the y and X datasets assuming that loan_status will be the label. Print the first 5 records of X
y = loans_df['loan_status']

X.head()
```

```python
# Question 22: Split the data using a 50/50 split and random seed = 12345
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=12345)

# Next, train a Decision Tree Classifier model using a random_state=12345 and fit the model with the appropriate data set
model = DecisionTreeClassifier(random_state=12345).fit(X_train, y_train)

# Print the Accuracy score using either the .accuracy_score() or .score() method of the Decision Tree Classifier model object. 
# Be sure to pass in the appropriate data into those objects. What is that score? Copy and paste the entire number with all decimals included
# model.score(X_test, y_test)
y_pred = model.predict(X_test)

y_test_dummies = pd.get_dummies(y_test, drop_first=True)
y_pred_dummies = pd.get_dummies(y_pred, drop_first=True)

# Accuracy  = (true positives + true negatives) / (total cases); ranges from 0 (worst) to 1 (best)
print(f"Accuracy:\t{metrics.accuracy_score(y_test, y_pred)}")

```

```python
# Question 23: Predict the labels for test dataset and display them in a dataframe along with the actual values. 
# Print only the LAST 10 records. Which row index was predicted incorrectly in these last 10 records? 
y_pred = model.predict(X_test)

df_results = pd.DataFrame({'Actual status' : y_test, 'Predicted status' : y_pred})
df_results.tail(10)
```
