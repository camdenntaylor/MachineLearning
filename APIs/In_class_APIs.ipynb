```python
import pandas as pd
import json
import requests

#      API                        /endpoint/ [specify parameters]
url = "https://api.coindesk.com/v1/bpi/currentprice.json"
req = requests.get(url)

req_json = req.json()
req_json
```

```python
for k, v in req_json.items():
  print(k, v)
```

```python
for entry in req_json['bpi']:
  print(req_json['bpi'][entry])
```

```python
# long way with complete customization options
df = pd.DataFrame(columns=['code', 'symbol', 'rate', 'description'])
# sets row index for df as 'code' column from json
df.set_index('code', inplace=True)

for key, value in req_json['bpi'].items():
  # Insert a row in df with each sub-dictionary data row
  df.loc[key] = [value['symbol'], value['rate_float'], value['description']]

df
```

```python
# medium way with a little less customization, but more convenience
df = pd.DataFrame(columns=['code', 'symbol', 'rate_float', 'description'])
df.set_index('code', inplace=True)

for key, value in req_json['bpi'].items():
  df.loc[key] = value

df.rename(columns={'rate_float' : 'rate'}, inplace=True)

df
```

```python
# fast way that eliminates iteration and pushes all customzation after df is created
df = pd.DataFrame(req_json['bpi'])
df = df.transpose()
df = df[['symbol', 'rate_float', 'description']]
df.rename(columns={'rate_float' : 'rate'})
```

```python
response = requests.get("https://datausa.io/api/data?drilldowns=Nation&measures=Population")

df = pd.DataFrame(columns=['ID Nation', 'Nation', 'Year', 'Population'])
df.set_index('Year', inplace=True)

for item in response.json()['data']:
  df.loc[item['Year']] = [item['ID Nation'], item['Nation'], item['Population']]

df

```
