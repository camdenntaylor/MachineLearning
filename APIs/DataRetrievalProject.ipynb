```python
### Pokemon Species (On my own project)
import json, requests, time, pandas as pd

df = pd.DataFrame(columns=['id', 'name', 'order', 'color', 'evolves_from', 'habitat', 'shape', 'generation', 'gender_rate', 
                           'capture_rate', 'growth_rate', 'hatch_counter', 'is_legendary', 'is_mythical', 'base_happiness'])
df.set_index('id', inplace=True)

for i in range(1, 100):
    url = f"https://pokeapi.co/api/v2/pokemon-species/{i}/"
    res = requests.get(url)
    res_json = json.loads(res.text)
    id = res_json['id']
    name = res_json['name']
    order = res_json['order']
    color = res_json['color']['name']

    #this try block handles first stage pokemon don't evolve from anything
    try:
        evolves_from = res_json['evolves_from_species']['name'] 
    except:
        evolves_from = 'Nothing'

    #accounts for null habitat values in later generations
    try:
        habitat = res_json['habitat']['name']
    except:
        habitat = 'None'

    shape = res_json['shape']['name']
    gender_rate = res_json['gender_rate']
    capture_rate = res_json['capture_rate']
    base_happiness = res_json['base_happiness']
    if res_json['is_legendary']:
        is_legendary = 1
    else:
        is_legendary = 0
  
    if res_json['is_mythical']:
        is_mythical = 1
    else:
        is_mythical = 0
     
    generation = res_json['generation']['name']
    hatch_counter = res_json['hatch_counter']
    if res_json['growth_rate']['name'] == 'slow':
        growth_rate = 1
    elif res_json['growth_rate']['name'] == 'medium-slow':
        growth_rate = 2
    elif res_json['growth_rate']['name'] == 'medium':
        growth_rate = 3
    elif res_json['growth_rate']['name'] == 'medium-fast':
        growth_rate = 4
    elif res_json['growth_rate']['name'] == 'fast':
        growth_rate = 5
    


    df.loc[id] = [name, order, color, evolves_from, habitat, shape, generation, gender_rate, 
                  capture_rate, growth_rate, hatch_counter, is_legendary, is_mythical, base_happiness]
    


    # time.sleep(2)

df.to_csv("pokemon_species.csv")
```
